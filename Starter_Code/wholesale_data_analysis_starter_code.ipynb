## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['subtotal'] = df['unit_price'] * df['qty']

print(df[['unit_price', 'qty', 'subtotal']].head(2))
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df['total_weight'] = df['unit_weight'] * df['qty']

df['shipping_price'] = df.apply(lambda row: row['total_weight'] * 7 if row['total_weight'] > 50 else row['total_weight'] * 10, axis = 1)

print(df[['unit_price','unit_weight', 'qty', 'total_weight', 'shipping_price']].head(3))

```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
# Calculate the sales tax
df['sales_tax'] = (df['subtotal'] + df['shipping_price']) * 0.0925

# Calculate the total price
df['total_price'] = df['subtotal'] + df['shipping_price'] + df['sales_tax']

# Display the first few rows to verify the new columns
print(df[['subtotal', 'shipping_price', 'sales_tax', 'total_price']].head(3))
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

# Calculate the cost of each line
df['line_cost'] = (df['unit_cost'] * df['qty']) + df['shipping_price']

# Display the first few rows to verify the new column
df.head(3)
```

```python
# Create a column for the profit of each line using line cost and line price
df['line_profit'] = df['total_price'] - df['line_cost']

df.head(3)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

# Calculate total spending for each client
top_5_spending = {}
for client_id in top_5_client_ids:
    client_df = df[df['client_id'] == int(client_id)]
    total_spending = (client_df['unit_price'] * client_df['qty']).sum()
    top_5_spending[client_id] = total_spending


print("Spending for the top 5 clients by quantity:")
for client_id, spending in top_5_spending.items():
    print(f"Client ID {client_id}: ${spending:.2f}")
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 

# Calculate the required totals for each client

summary_data = []
for client_id in top_5_client_ids:
    client_df = df[df['client_id'] == int(client_id)]
    
    total_units = client_df['qty'].sum()
    total_cost = client_df['unit_cost'] * client_df['qty']
    total_shipping = client_df['shipping_price'].sum()
    total_revenue = (client_df['unit_price'] * client_df['qty']).sum()
    total_profit = (total_revenue - total_cost - total_shipping).sum()
    
    summary_data.append({
        'client_id': client_id,
        'total_units': total_units,
        'total_shipping': total_shipping,
        'total_revenue': total_revenue,
        'total_profit': total_profit
    })

# Create the summary DataFrame
summary_df = pd.DataFrame(summary_data)

# Set the client_id as the index
summary_df.set_index('client_id', inplace=True)

# Display the summary DataFrame
print(summary_df)
```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 
money_columns = ['total_shipping', 'total_revenue', 'total_profit']

# Define a function that converts a dollar amount to millions.
def currency_format_millions(value):
    return f"${value / 1000000:.2f}M"

# Apply the currency_format_millions function to only the money columns. 
for col in money_columns:
    summary_df[col] = summary_df[col].apply(currency_format_millions)


# Rename the columns to reflect the change in the money format. 
summary_df = summary_df.rename(columns={'client_id': 'Client ID', 
                                        'total_units': 'Total Units', 
                                        'total_shipping': 'Shipping (Millions)', 
                                        'total_revenue': 'Total Revenue (Millions)', 
                                        'total_profit': 'Total Profit (Millions)'
                                        })

# Set the Client ID as the index
#summary_df.set_index('Client ID', inplace=True)

# Display the formatted summary DataFrame
print(summary_df)

```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
sorted_df = summary_df.sort_values(by='Total Profit (Millions)', ascending=False)
sorted_df
```
